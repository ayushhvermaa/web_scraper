```python
# This Program takes the input of the University CalLink and outputs a dictionary containing the list of school clubs,
# club links, club status, and primary contact email

#Below are the Python documentation and imports used

import requests
import pandas as pd
import json
from urllib import request
from bs4 import BeautifulSoup
import json
from urllib.request import Request, urlopen
```

```python
# Using the CalLink URL, input the link into an application known as Postman. In Postman, click the "+" in order to 
# open a new session, paste the URL and send it. Make sure the settings are on Pretty and set to JSON to obtain file
# Then, create folder called "*University Name* Data" and paste in Desktop, and using the terminal call cd Desktop
# and then calling "touch *filename*.json" and drag the created json into the folder. Paste the contents of the Postman
# data into the empty JSON and load below into our code
```

```python
with open('berkeleydata.json') as f:
   data = json.load(f)

print(data)
```

```python
#Below we can pull the values of and simplify our data
```

```python
orgs = data['value']
orgs[0]
```

```python
#This Next code is meant to pull the name, website key, description, and status
```

```python
parsed_orgs = list(map(lambda x: dict(name=x['Name'], web=x['WebsiteKey'], desc=x['Description'], status=x['Status']), orgs))
```

```python
parsed_orgs[0]
```

```python
# Following function uses any of the keys from above to access the CalLink URL and pull email from metadata
```

```python
def email_getter(key):
    link = 'https://callink.berkeley.edu/organization/' + key
    site = urlopen(Request(link,headers={'User-Agent': 'Mozilla/5.0'})).read()
    soupsite = BeautifulSoup(site, 'html.parser')
    return json.loads(soupsite.find_all("script")[5].text[25:-1])['preFetchedData']['organization']['email']
    
```

```python
# Create empty array
```

```python
berkeleyData = []
```

```python
# Function maps parsed orgs and keeps all previous information, and also sifts through all the keys inputting them 
# into email getter and adding them to total dictionary
```

```python
def getAllData(start, end):  

    return list(map(lambda x: dict(name=x['name'], web=x['web'], desc=x['desc'], status=x['status'], email=email_getter(x['web'])),parsed_orgs[start:end]))
```

```python
#Append function call below, getAllData start and stop parameters are taking start to end th of total orgs. 
```

```python
berkeleyData.append(getAllData(0,100))
berkeleyData
```
